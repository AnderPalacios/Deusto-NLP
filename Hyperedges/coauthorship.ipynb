```python
import re
import pandas as pd

def parse_authors(author_str):
    if not isinstance(author_str, str) or author_str.strip() == "":
        return []
    
    authors = []
    
    # Step 1: Split by ' and ' (primary delimiter)
    parts = re.split(r'\s+and\s+', author_str)
    
    for part in parts:
        part = part.strip()
        if not part:
            continue
        
        # Step 2: Handle 'Lastname, F.' by swapping to 'F. Lastname'
        comma_match = re.match(r'^(.+?),\s*(.+)$', part)
        if comma_match:
            last, first = comma_match.groups()
            author_name = f"{first} {last}".strip()
            authors.append(author_name)
        else:
            # Step 3: Keep multi-word names intact
            authors.append(part)
    
    return authors
```

```python
cora['Author_list'] = cora['Author'].apply(parse_authors)
```
